## 4.5 ROC Curves 
(receiver operating characteristics)

### TPR and FPR

```python
tpr = tp / (tp+fn)
tpr
```

```python
fpr = fp / (fp+tn)
fpr
```

```python
scores = []

thresholds = np.linspace(0, 1, 101)

for t in thresholds:

    actual_positive = (y_val == 1)
    actual_negative = (y_val == 0)
    predict_positive = (y_pred >= t)
    predict_negative = (y_pred < t)

    tp = (predict_positive & actual_positive).sum()
    tn = (predict_negative & actual_negative).sum()
    fp = (predict_positive & actual_negative).sum()
    fn = (predict_negative & actual_positive).sum()

    scores.append((t, tp, fp, fn, tn))
```

```python
columns=['threshold', 'tp', 'fp', 'fn', 'tn']
df_scores = pd.DataFrame(scores, columns=columns)
df_scores
```

```python
df_scores[::10]
```

```python
df_scores['tpr'] = df_scores.tp / (df_scores.tp + df_scores.fn)
df_scores['fpr'] = df_scores.fp / (df_scores.fp + df_scores.tn)
```

```python
plt.plot(thresholds, df_scores.tpr, label='tpr')
plt.plot(thresholds, df_scores.fpr, label='fpr')

plt.xlabel('threshold')
plt.legend()
```

### Random model

```python
np.random.seed(1)
y_rand = np.random.uniform(0, 1, size=len(y_val))
```

```python
((y_rand >= 0.5) == y_val).mean()
```

```python
def tpr_fpr_dataframe(y_val , y_pred):
    scores = []

    thresholds = np.linspace(0, 1, 101)

    for t in thresholds:

        actual_positive = (y_val == 1)
        actual_negative = (y_val == 0)
        predict_positive = (y_pred >= t)
        predict_negative = (y_pred < t)

        tp = (predict_positive & actual_positive).sum()
        tn = (predict_negative & actual_negative).sum()
        fp = (predict_positive & actual_negative).sum()
        fn = (predict_negative & actual_positive).sum()

        scores.append((t, tp, fp, fn, tn))

    columns=['threshold', 'tp', 'fp', 'fn', 'tn']
    df_scores = pd.DataFrame(scores, columns=columns)

    df_scores['tpr'] = df_scores.tp / (df_scores.tp + df_scores.fn)
    df_scores['fpr'] = df_scores.fp / (df_scores.fp + df_scores.tn)

    return df_scores
```

```python
df_rand = tpr_fpr_dataframe(y_val, y_rand)
df_rand[::10]
```

```python
plt.plot(thresholds, df_rand.tpr, label='tpr')
plt.plot(thresholds, df_rand.fpr, label='fpr')

plt.xlabel('threshold')
plt.legend()
```

### Ideal model

```python
num_neg = (y_val==0).sum()
num_pos = (y_val==1).sum()
num_neg, num_pos
```

```python
y_ideal = np.repeat([0, 1], [num_neg, num_pos])
y_ideal
```

```python
y_ideal_pred = np.linspace(0, 1, len(y_val))
```

```python
((y_ideal_pred >= 0.726) == y_ideal).mean()
```

```python
df_ideal = tpr_fpr_dataframe(y_ideal, y_ideal_pred)
df_ideal[::10]
```

```python
plt.plot(thresholds, df_ideal.tpr, label='tpr')
plt.plot(thresholds, df_ideal.fpr, label='fpr')

plt.xlabel('threshold')
plt.legend()
plt.title('Ideal Model')
```

```python
plt.figure(figsize=(6,6))

plt.plot(df_scores.fpr, df_scores.tpr, label='model') # a good model will stay close to the ideal line
plt.plot(df_rand.fpr, df_rand.tpr, label='random')
plt.plot(df_ideal.fpr, df_ideal.tpr, label='ideal') # top left corner (TPR 1 FPR 0 @ threshold = 0.726)

plt.ylabel('TPR')
plt.xlabel('FPR')
plt.title('ROC curve')
plt.legend()
```
